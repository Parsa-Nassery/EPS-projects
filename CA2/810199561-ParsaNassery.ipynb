```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
import pandas as pd
import seaborn as sns


def find_average(nums):
    sum = 0
    length = len(nums)
    for i in range(0, length):
        sum = sum + nums[i]
    return sum / length

def find_variance(nums):
    length = len(nums)
    ave = find_average(nums)
    var = 0
    for i in range(0, length):
        var = var + round(((nums[i] - ave) ** 2), 4)
    var = var / length
    return var

def find_correlation_coefficient(X, Y):
    up = 0
    down_x = 0
    down_y = 0
    total_down = 0
    ave_X = find_average(X)
    ave_Y = find_average(Y)
    for i in range(0, len(X)):
        temp = (X[i] - ave_X) * (Y[i] - ave_Y)
        up = up + temp
        down_x = down_x + (X[i] - ave_X) ** 2
        down_y = down_y + (Y[i] - ave_Y) ** 2
    total_down = np.sqrt(down_x * down_y)
    r = up / total_down
    return r



csv = pd.read_csv('StudentsPerformance.csv')
df = pd.DataFrame(csv)
grades1 = list(df.iloc[:, 14])
grades2 = list(df.iloc[:, 15])
print("correlation coefficient of Grades 1 and 2 is :", find_correlation_coefficient(grades1, grades2))

        
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
import pandas as pd
import seaborn as sns
 

def find_average(nums):
    sum = 0
    length = len(nums)
    for i in range(0, length):
        sum = sum + nums[i]
    return sum / length


csv = pd.read_csv('grades.csv')
df = pd.DataFrame(csv)
parent_class = list(df.iloc[:, 3])
math_score = list(df.iloc[:, 6])
reading_score = list(df.iloc[:, 7])
writing_score = list(df.iloc[:, 8])\

total_grades = []
for i in range(0 ,len(writing_score)):
    total_grades.append((math_score[i] + reading_score[i] + writing_score[i]) / 3)

bachelors_degree = []
some_college = []
masters_degree = []
associates_degree = []
highschool = []
some_high_school = []

for i in range(0, len(writing_score)):
    if parent_class[i] == "bachelor's degree":
        bachelors_degree.append(total_grades[i])
    if parent_class[i] == "some college":
        some_college.append(total_grades[i])
    if parent_class[i] == "master's degree":
        masters_degree.append(total_grades[i])
    if parent_class[i] == "associate's degree":
        associates_degree.append(total_grades[i])
    if parent_class[i] == "high school":
        highschool.append(total_grades[i])
    if parent_class[i] == "some high school":
        some_high_school.append(total_grades[i])

bachelors_degree = find_average(bachelors_degree)
some_college = find_average(some_college)
some_high_school = find_average(some_high_school)
masters_degree = find_average(masters_degree)
associates_degree = find_average(associates_degree)
highschool = find_average(highschool)

result = [bachelors_degree, some_college, some_high_school, masters_degree, associates_degree, highschool]

n = np.arange(0, 6, 1)
plt.figure(figsize=(15,8))
plt.bar(n, result, edgecolor='grey',  color ='b')
plt.title("Grade per parents")
plt.xlabel("Parents")
plt.ylabel("Grade")
plt.xticks([0,1,2,3,4,5], ['bachelors_degree', 'some_college', 'some_high_school', 'masters_degree', 'associates_degree', 'highschool'])
plt.show()

```
