```python
import re
import numpy as np
import matplotlib.pyplot as plt
import random

T_res = []
F_res = []


def remove_and_choose_new_door(correct, choice):
    if correct == choice:
        temp_door = np.random.randint(1,4,(1,1))
        while temp_door[0][0] == correct:
            temp_door = np.random.randint(1,4,(1,1))
        if correct != 1 and temp_door[0][0] != 1:
            return 1
        if correct != 2 and temp_door[0][0] != 2:
            return 2
        return 3
    else:
        return correct



def newgame():
    Tcount = 0
    Fcount = 0
    for i in range(0, 1000):
        correct_door = np.random.randint(1,4,(1,1))
        first_choice = np.random.randint(1,4,(1,1))
        final_choice = remove_and_choose_new_door(correct_door[0][0], first_choice[0][0])
        if correct_door[0][0] == final_choice:
            Tcount = Tcount + 1
        else:
            Fcount = Fcount + 1
        T_res.append(Tcount)
        F_res.append(Fcount)

newgame()

T_res = np.array(T_res)
F_res = np.array(F_res)

x = np.arange(0, 1000, 1)

plt.plot(x, T_res, 'g', label = 'Win')
plt.plot(x, F_res, 'r', label = 'Lose')
plt.legend(loc = 'upper right')
plt.title('Goat Game!!')
plt.xlabel('Game Number')
plt.ylabel('Game Result')
plt.show()
```

```python
from math import factorial
import numpy as np
import matplotlib.pyplot as plt
import random



def C(choose, From):
    res = factorial(From)/(factorial(choose)*factorial(From-choose))
    return res

def minimum_6n(n):
    res = 1
    for i in range(0, n):
        temp = C(i, 6 * n)
        temp = temp / (6 ** i)
        temp = temp * ((5/6) ** (6 * n - i))
        res = res - temp
    return res

results = []
for i in range(0, 200):
    results.append(minimum_6n(i))

results = np.array(results)

x = np.arange(0, 200, 1)

plt.plot(x, results, 'b', label = 'Probability')
plt.legend(loc = 'upper right')
plt.title('Newton - Pepys')
plt.xlabel('n')
plt.ylabel('Probability')
plt.show()
```
